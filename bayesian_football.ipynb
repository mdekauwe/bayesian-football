```python
import pandas as pd
from pandas import DataFrame
import pymc
```

```python
#data source: http://en.wikipedia.org/wiki/1991%E2%80%9392_Serie_A
df = pd.read_csv('https://raw.github.com/DanielWeitzenfeld/bayesian-football/master/seria_a_1991.txt', sep='\t', index_col=0)
rows = []
for i in df.index:
    for c in df.columns:
        if i == c: continue
        score = df.ix[i, c]
        score = [int(row) for row in score.split('-')]
        rows.append([i, c, score[0], score[1]])
df = DataFrame(rows, columns = ['home', 'away', 'home_score', 'away_score'])
teams = df.home.unique()
teams = DataFrame(teams, columns=['team'])
teams['i'] = teams.index
df = pd.merge(df, teams, left_on='home', right_on='team', how='left')
df.rename(columns = {'i': 'i_home'}, inplace=True)
df = pd.merge(df, teams, left_on='away', right_on='team', how='left')
df.rename(columns = {'i': 'i_away'}, inplace=True)
df = df.drop('team_x', 1)
df = df.drop('team_y', 1)
```
